# SCC RMSE

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
```

# Read SCC

```python
files=np.zeros(4)

path1='SCC_maple_48/'
path2='SCC_maple_144/'
path3='SCC_com/'

path4='rmse_maple_48/'
path5='rmse_maple_144/'
path6='rmse_com/'


files1=os.listdir(path1)
files2=os.listdir(path2)
files3=os.listdir(path3)

files4=os.listdir(path4)
files5=os.listdir(path5)
files6=os.listdir(path6)

scc_maple48=np.zeros((len(files1),37))
scc_maple144=np.zeros((len(files2),37))
scc_com=np.zeros((len(files3),37))


for i in range(0,len(files1)):
    f=open(path1+'/'+files1[i])
    data=f.readlines()
    a=0
    for line in data:
        scc_maple48[i,a] =line
        a +=1

for i in range(0,len(files2)):
    f=open(path2+'/'+files2[i])
    data=f.readlines()
    a=0
    for line in data:
        scc_maple144[i,a] =line
        a +=1        
for i in range(0,len(files3)):
    f=open(path3+'/'+files3[i])
    data=f.readlines()
    a=0
    for line in data:
        scc_com[i,a] =line
        a +=1
        
avg_scc_maple48=np.average(scc_maple48,axis=0)
avg_scc_maple144=np.average(scc_maple144,axis=0)
avg_scc_com=np.average(scc_com,axis=0)
```

# Read RMSE
        

```python
rmse_maple48=np.zeros((len(files4),37))
rmse_maple144=np.zeros((len(files5),37))
rmse_com=np.zeros((len(files6),37))


for i in range(0,len(files4)):
    f=open(path4+'/'+files4[i])
    data=f.readlines()
    a=0
    for line in data:
        rmse_maple48[i,a] =line
        a +=1

for i in range(0,len(files5)):
    f=open(path5+'/'+files5[i])
    data=f.readlines()
    a=0
    for line in data:
        rmse_maple144[i,a] =line
        a +=1        
for i in range(0,len(files6)):
    f=open(path6+'/'+files6[i])
    data=f.readlines()
    a=0
    for line in data:
        rmse_com[i,a] =line
        a +=1
        
avg_rmse_maple48=np.average(rmse_maple48,axis=0)
avg_rmse_maple144=np.average(rmse_maple144,axis=0)
avg_rmse_com=np.average(rmse_com,axis=0)
```

# Plot

```python
fig = plt.figure()
ax1=fig.add_subplot(111)
ax1.plot(avg_scc_maple48,'b',label='MAPLE48x48',linewidth=3)
ax1.plot(avg_scc_maple144,'g',label='MAPLE144x144',linewidth=3)
ax1.plot(avg_scc_com,'r',label='MAPLE_WRF',linewidth=3)
ax1.set_ylabel('SCC',size=20)
ax1.set_xticklabels(['0','1','2','3','4'])
ax1.set_xticks([0,6,12,18,24])
ax1.set_xlim(0,24)
ax1.set_ylim(0.4,1)

x=np.mgrid[slice(0,25,1)]
ax2=ax1.twinx()
ax2.set_ylabel('RMSE(dBZ)',size=20)
ax2.set_ylim(0,12)
ax2.plot(avg_rmse_maple48,'b--',linewidth=3)
ax2.plot(avg_rmse_maple144,'g--',linewidth=3)
ax2.plot(avg_rmse_com,'r--',linewidth=3)
ax1.set_xlabel('Hour',size=20)
ax1.set_title('16 Typhoon Cases',size=25)
ax1.legend(loc=2,bbox_to_anchor=(0.2,0.35),fontsize='xx-large')
ax1.grid()
fig.set_size_inches(8,6)
fig.savefig('/home/miayao/average/16_typhoon_scc_rmse_echo.png',dpi=600)
```
