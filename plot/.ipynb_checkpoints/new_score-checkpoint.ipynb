# New Score
## pod, far, bias, csi

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
stdfar=np.zeros((100,100),dtype=float)
stdpod=np.zeros((100,100),dtype=float)
stdbias=np.zeros((100,100),dtype=float)
stdcsi=np.zeros((100,100),dtype=float)
for i in range(0,100):
    stdfar[:,i]=np.linspace(0,1,100)
    stdpod[i,:]=np.linspace(1,0,100)
for i in range(0,100):
    for j in range(0,100):
        if (stdfar[i,j] == 1. and stdpod[i,j] != 0.) :
            stdbias[i,j]=np.nan
        elif (stdfar[i,j] == 1 and stdpod[i,j] == 0):
            stdbias[i,j]=0.
        else:
            stdbias[i,j]=(stdpod[i,j])/(1-stdfar[i,j])
        if (stdfar[i,j] == 1 or stdpod[i,j] == 0):
            stdcsi[i,j]=np.nan
        else:
            stdcsi[i,j]=(1./((1/stdpod[i,j])+1/(1-stdfar[i,j])-1))
```

## Read data

```python
files=np.zeros(4)

path1='pod/maple48/'
path2='pod/maple144/'
path3='pod/com/'

path4='far/maple48/'
path5='far/maple144/'
path6='far/com/'


files1=os.listdir(path1)
files2=os.listdir(path2)
files3=os.listdir(path3)

files4=os.listdir(path4)
files5=os.listdir(path5)
files6=os.listdir(path6)

pod_maple48=np.zeros((len(files1),37))
pod_maple144=np.zeros((len(files2),37))
pod_com=np.zeros((len(files3),37))

far_maple48=np.zeros((len(files4),37))
far_maple144=np.zeros((len(files5),37))
far_com=np.zeros((len(files6),37))

for i in range(0,len(files1)):
    f=open(path1+'/'+files1[i])
    data=f.readlines()
    a=0
    for line in data:
        if float(line) == -999.:
            pod_maple48[i,a]=np.nan
        else:
            pod_maple48[i,a]=line
        a +=1
for i in range(0,len(files2)):
    f=open(path2+'/'+files2[i])
    data=f.readlines()
    a=0
    for line in data:
        if float(line) == -999.:
            pod_maple144[i,a]=np.nan
        else:
            pod_maple144[i,a]=line
        a +=1
    
for i in range(0,len(files3)):
    f=open(path3+'/'+files3[i])
    data=f.readlines()
    a=0
    for line in data:
        if float(line) == -999.:
            pod_com[i,a]=np.nan
        else:
            pod_com[i,a]=line
        a +=1

for i in range(0,len(files4)):
    f=open(path4+'/'+files4[i])
    data=f.readlines()
    a=0
    for line in data:
        if float(line) == -999.:
            far_maple48[i,a]=np.nan
        else:
            far_maple48[i,a]=line
        a +=1
        
for i in range(0,len(files5)):
    f=open(path5+'/'+files5[i])
    data=f.readlines()
    a=0
    for line in data:
        if float(line) == -999.:
            far_maple144[i,a]=np.nan
        else:
            far_maple144[i,a]=line
        a +=1     
for i in range(0,len(files6)):
    f=open(path6+'/'+files6[i])
    data=f.readlines()
    a=0
    for line in data:
        if float(line) == -999.:
            far_com[i,a]=np.nan
        else:
            far_com[i,a]=line
        a +=1
```

# 矩陣處理

```python
avg_pod_maple48=np.nanmean(pod_maple48,axis=0)
avg_pod_maple144=np.nanmean(pod_maple144,axis=0)
avg_pod_com=np.nanmean(pod_com,axis=0)

avg_far_maple48=np.nanmean(far_maple48,axis=0)
avg_far_maple144=np.nanmean(far_maple144,axis=0)
avg_far_com=np.nanmean(far_com,axis=0)
```

# Average

```python
avg_pod_maple48=np.array(avg_pod_maple48,dtype=float)
avg_far_maple48=np.array(avg_far_maple48,dtype=float)

avg_pod_maple144=np.array(avg_pod_maple144,dtype=float)
avg_far_maple144=np.array(avg_far_maple144,dtype=float)

avg_pod_com=np.array(avg_pod_com,dtype=float)
avg_far_com=np.array(avg_far_com,dtype=float)
```

# Plot

```python
fig,ax = plt.subplots()

ax.plot(1-avg_far_maple48[:25:3],avg_pod_maple48[:25:3],'bo-',label='MAPLE48x48',linewidth=5,markersize=9)
ax.plot(1-avg_far_maple144[:25:3],avg_pod_maple144[:25:3],'go--',label='MAPLE144x144',linewidth=3,markersize=6)
ax.plot(1-avg_far_com[:25:3],avg_pod_com[:25:3],'ro:',label='MAPLE_WRF',linewidth=3,markersize=7)


ax.set_ylabel('Probability of Detection (POD)',size=30)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_xlabel('Success Ratio (1-FAR)',size=30)

CS1=ax.contour(1-stdfar,stdpod,stdbias,[0.1,0.3,0.5,0.7,0.8,1,1.2,1.5,2,3,10],colors='k',alpha=0.5,linestyles='dashed')
ax.clabel(CS1,inline=True,fontsize=15,fmt='%1.1f')
CS2=ax.contourf(1-stdfar,stdpod,stdcsi,cmap="coolwarm",alpha=0.4)
cbar=fig.colorbar(CS2,ticks=[0,0.15,0.3,0.45,0.6,0.75,0.9])
cbar.values=(0,1)
cbar.set_label('CSI',size=25)
fig.set_size_inches(12,9)
ax.legend(loc=2,bbox_to_anchor=(0.55,0.25),fontsize='xx-large')
ax.set_title('16 Typhoon Cases',size=40)

fig.savefig('/home/miayao/average/16cases_newscore_echo_15dBZ_except_MAPLE_WRF.png',dpi=600)
```
